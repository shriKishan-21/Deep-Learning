```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Embedding,SimpleRNN,LSTM,GRU,Bidirectional
```

```python
num_words = 10000
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=num_words)
```

```python
max_len = 100
x_train = pad_sequences(x_train,maxlen=max_len,padding='post',truncating='post')
x_test = pad_sequences(x_test,maxlen=max_len,padding='post',truncating='post')
```

```python
embedding_dim = 32
```

```python
# Simple RNN
```

```python
model = Sequential([
    Embedding(input_dim=num_words,output_dim=embedding_dim,input_length=max_len),
    SimpleRNN(5),
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train,epochs=5,validation_split=0.2)
```

```python
# Bidirectional RNN
```

```python
model2 = Sequential([
    Embedding(output_dim=embedding_dim,input_length=max_len,input_dim=num_words),
    Bidirectional(SimpleRNN(5)),
    Dense(1,activation='sigmoid')
])

model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model2.summary()
```

```python
history2 = model2.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
# Since the Deep RNN is applicable for both GRU and LSTM 
# Likewise Bidirectional RNN are applicable for both GRU and LSTM
```

```python
# Bidirectional LSTM
```

```python
model3 = Sequential([
    Embedding(output_dim=embedding_dim,input_length=max_len,input_dim=num_words),
    Bidirectional(LSTM(5)),
    Dense(1,activation='sigmoid')
])

model3.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model3.summary()
```

```python
history3 = model3.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
# Bidirectional GRU
```

```python
model4 = Sequential([
    Embedding(output_dim=embedding_dim,input_length=max_len,input_dim=num_words),
    Bidirectional(GRU(5)),
    Dense(1,activation='sigmoid')
])

model4.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model4.summary()
```

```python
history4 = model4.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```
