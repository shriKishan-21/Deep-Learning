```python
import numpy as np
import pandas as pd

docs = [
    'go india',
    'india india',
    'hip hip hurray',
    'jeetega bhai jeetega india jeetega',
    'bharat mata ki jai',
    'kohli kohli',
    'sachin sachin',
    'dhoni dhonni',
    'modi ji ke jai',
    'inquilab zindabad'
]
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(oov_token='Error')
# oov_token = out of vocabulary token -> that is if there is any work that came and is not present in ou vocab it will provide with oob_token value/
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequence = tokenizer.texts_to_sequences(docs)
sequence
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
sequence = pad_sequences(sequence,padding='post')
```
