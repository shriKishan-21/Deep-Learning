```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
```

```python
img = image.load_img('cat.jpg',target_size=(200,200))
```

```python
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
type(img)
# Here the image is PIL - Pyhton imagining library object
# Not a numpy array
```

```python
data_gen = ImageDataGenerator(
    rotation_range=30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = 'nearest'
)

# here fill_mode will fill the spaces that are kept blank by shifting of image according to the type passed
# nerest - fill by filling nearest image 
# reflect - fill by exactly what was before
# constant - do not fill keep as it is
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i=0

for output in data_gen.flow(input_batch,batch_size=1,save_to_dir='aug'):
    i = i + 1
    if i == 10:
        break
```

```python
# flow we use when working with a single numpy image
# flow_from_directory is used when multiple images are provided in a directory
```
