```python
import cv2
import os
import random
import numpy as np                   
```

```python
# Need to create same dataset as ImageData but with 1000 images only
# so instead of creating it normally lets use this code
import shutil

random.seed(42)

source = "ImageData"
destination = "ImageDataSmall"

for split in ["test","train"]:
    for cls in ["cats","dogs"]:
        src_folder = os.path.join(source,split,cls)
        dst_folder = os.path.join(destination,split,cls)
        
        os.makedirs(dst_folder,exist_ok=True)
        
        image = os.listdir(src_folder)
        selected = random.sample(image,1000)
        
        for img in selected:
            shutil.copy(os.path.join(src_folder,img),os.path.join(dst_folder,img))
            
print("Success")
```

```python
mydir = 'ImageDataSmall/train'
```

```python
categories = ['cats','dogs']
```

```python
data = []

for i in categories:
    folder_path = os.path.join(mydir,i)
    
    if i == 'cats':
        label = 0
    else:
        lable = 1
        
    for j in os.listdir(folder_path):
        img_path = os.path.join(folder_path,j)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(150,150))
        data.append([img,label])
```

```python
random.shuffle(data)
```

```python
x = []
y = []

for i in data:
    x.append(i[0])
    y.append(i[1])
```

```python
y = np.array(y)
x = np.array(x)
```

```python
y.shape
```

```python
x = x/255
```

```python
# Create CNN from here

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Activation,Dropout
```

```python
model = Sequential()

model.add(Conv2D(32,(3,3),input_shape=(150,150,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())

model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer='rmsprop',metrics=['accuracy'],loss='binary_crossentropy')
```

```python
model.fit(x,y,epochs=5,validation_split=0.1)
```
