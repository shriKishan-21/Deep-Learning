```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
```

```python
batch_size = 16

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

# this is the augmentation configuration we will use for testing
train_Datagen = ImageDataGenerator(rescale=1./255)

# this is generator that will read images found in subfolder of 'data/train' and idefinitely generate batches of augmented image data
train_generator = train_datagen.flow_from_directory(
    'ImageDataSmall/train', # target directory
    target_size = (150,150), # all image resize to 150,150
    batch_size = batch_size,
    class_mode = 'binary' # since we use binary _crossentropy loss, we need binary labels
)

# generator for validation
validation_generator = train_datagen.flow_from_directory(
    'ImageDataSmall/valid', # target directory
    target_size = (150,150), # all image resize to 150,150
    batch_size = batch_size,
    class_mode = 'binary' # since we use binary _crossentropy loss, we need binary labels
)
```

```python
# Create CNN from here

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Activation,Dropout
```

```python
model = Sequential()

model.add(Conv2D(32,(3,3),input_shape=(150,150,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())

model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer='rmsprop',metrics=['accuracy'],loss='binary_crossentropy')
```

```python
# Note : Whenever we use imageGenerator we will use fit_generator instead of fit
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=2000 // batch_size,
    epochs = 25,
    validation_data = validation_generator,
    validation_steps = 800 // batch_size
)
```
