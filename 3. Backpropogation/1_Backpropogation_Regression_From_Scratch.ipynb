```python
# Here we take W = 0.1 & b = 0
# This function initializes weights and bias
```

```python
def initialize_parameters(layer_dims):
    
    np.random.seed(3)
    parameters = {}
    # No of layers
    L = len(layer_dims)
    
    # loop to calculate number of weights and bias
    for l in range(1,L):
        parameters['W'+str(l)] = np.ones((layer_dims[l-1],layer_dims[l]))*0.1
        parameters['b'+str(l)] = np.zeros((layer_dims[l],1))
        
    return parameters
```

```python
initialize_parameters([2,2,1])
```

```python
# Using formula z = w.Tx+b
# w = weights
# x = previous nodes value
# b = bias
# below A_prev = x
```

```python
def linear_forward(A_prev ,W ,b):
    z = np.dot(W.T,A_prev) + b
    
    return z
```

```python
# Forward Propogation
def L_layer_forward(x,parameters):
    A = x
    L = len(parameters) // 2  # Number of layers in neural network
    
    for l in range(1,L+1):
        A_prev = A
        W1 = parameters['W'+str(l)]
        b1 = parameters['b'+str(l)]
        print("A"+str(l-1)+": ",A_prev)
        print("W"+str(l)+": ",W1)
        print("b"+str(l)+": ",b1)
        print("--"*20)
        
        A = linear_forward(A_prev,W1,b1)
        print("A"+str(l)+": ",A)
        print("--"*20)
        
    return A,A_prev
    
```

```python
# The above code is for prediction on data
# lets take an exapmle by taking first student data
# Step 1 : Select 1 row(random)
x = df[['cgpa','profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]
x
```

```python
# Parameters initialize - same as we provide above
parameters = initialize_parameters([2,2,1])

# Step 2 : Predict using forward propogation
y_hat,A1 = L_layer_forward(x,parameters)
```

```python
A1
```

```python
# Loss for first student - (y-y_hat)^2
# Step 3 : Calculate loss
loss = (y-y_hat[0][0])**2
loss
```

```python
# Step 4: Update weights and bias using gd
```

```python
def update_parameters(parameters,y,y_hat,A1,x):
    # param_new = param_old - lr * gradient
    parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y-y_hat) * A1[0][0])
    parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y-y_hat) * A1[1][0])
    parameters['b2'][0][0] = parameters['b2'][0][0] + (0.001 * 2 * (y-y_hat))
    
    parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y-y_hat) * parameters['W2'][0][0] * x[0][0])
    parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y-y_hat) * parameters['W2'][0][0] * x[1][0])
    parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y-y_hat) * parameters['W2'][0][0])
    
    parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y-y_hat) * parameters['W2'][1][0] * x[0][0])
    parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y-y_hat) * parameters['W2'][1][0] * x[1][0])
    parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y-y_hat) * parameters['W2'][1][0])
    
    
    
```

```python
update_parameters(parameters,y,y_hat[0][0],A1,x)
```

```python
# Forward Propogation
def L_layer_forward(x,parameters):
    A = x
    L = len(parameters) // 2  # Number of layers in neural network
    
    for l in range(1,L+1):
        A_prev = A
        W1 = parameters['W'+str(l)]
        b1 = parameters['b'+str(l)]
        # print("A"+str(l-1)+": ",A_prev)
        # print("W"+str(l)+": ",W1)
        # print("b"+str(l)+": ",b1)
        # print("--"*20)
        
        A = linear_forward(A_prev,W1,b1)
        # print("A"+str(l)+": ",A)
        # print("--"*20)
        
    return A,A_prev
```

```python
# Epoch implementation

parameters = initialize_parameters([2,2,1])
epochs = 7

for i in range(epochs):
    Loss = []
    
    for j in range(df.shape[0]):
        
        x = df[['cgpa','profile_score']].values[j].reshape(2,1)
        y = df[['lpa']].values[j][0]
        
        # parameters initialization
        
        y_hat , A1 = L_layer_forward(x,parameters)
        y_hat = y_hat[0][0]
        
        # update parameters
        update_parameters(parameters,y,y_hat,A1,x)
        
        Loss.append((y-y_hat)**2)
    
    print('Epoch -',i+1,'Loss - ',np.array(Loss).mean())
    
parameters
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(2,activation='linear',input_dim=2))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
# Keras initialize weights and bias randomly
# to make it similar to what we did above we'll create new weights and bias 
```

```python
new_weights = [np.array([[0.1 ,0.1],
                        [0.1,0.1]],dtype=np.float32),
                np.array([0. , 0.],dtype=np.float32),
                np.array([[0.1],
                          [0.1]],dtype=np.float32),
                np.array([0.],dtype=np.float32)
               ]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['lpa'].values,epochs=75,verbose=1,batch_size=1)
```
