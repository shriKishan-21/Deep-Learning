```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DailyDelhiClimate.csv',usecols=['date','meantemp'],nrows=114)
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
# To calculate the exponentially weighted moving average, we can use the `ewm` method in pandas.
x1 = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
x1
```

```python
df['ewma'] = x1
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black',label='alpha = 0.9')
plt.legend()
plt.show()
```
