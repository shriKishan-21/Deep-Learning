```python
from sklearn.datasets import make_classification
import numpy as np
x,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,
                          random_state=41,hypercube=False,class_sep=15)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(x,y):
    w1=w2=b=1
    lr = 0.1
    
    for j in range(100):
        for i in range(x.shape[0]):
            
            # check condition
            z = w1*x[i][0] + w2*x[i][1] + b
            
            # differentiation of L wrt w1 ,w2,b checks the same condition
            #  0 if yi*f(xi) >=0 -> no change so we'll see second condition
            # update if yi*f(xi)<0
            if z*y[i]<0:
                w1 = w1 + lr*y[i]*x[i][0]
                w2 = w2 + lr*y[i]*x[i][1]
                b = b + lr*y[i]
                
    return w1,w2,b
            
```

```python
w1,w2,b = perceptron(x,y)
```

```python
print("w1:", w1)
print("w2:", w2)
print("b:", b)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=2)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
